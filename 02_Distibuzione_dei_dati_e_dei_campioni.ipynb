# **Campionamento e Bias**

**Campione**: Un sottoinsieme di un insieme di dati più ampio.

**Popolazione**: Il insieme di dati più ampio o l'idea di un insieme di dati.

**N (n)**: La dimensione della popolazione (campione).

**Campionamento casuale**: Il processo di estrazione di elementi casuali in un campione.

**Campionamento stratificato**: Il processo di divisione della popolazione in strati e di campionamento casuale da ciascun strato.

**Strato (pl., strati)**: Un sottogruppo omogeneo della popolazione con caratteristiche comuni.

**Campione casuale semplice**: Il campione che risulta dal campionamento casuale senza stratificare la popolazione.

**Bias**: Errore sistematico.

**Bias campionario**: Un campione che non rappresenta correttamente la popolazione.




## **Campionamento casuale e bias**

il bias è un errore sistematico che si verifica quando le misurazioni o le osservazioni non sono rappresentative dell'intera popolazione.

La qualità dei dati è spesso più importante della quantità di dati. Dati di alta qualità sono completi, consistenti, puliti e accurati. Il campionamento casuale può aiutare a migliorare la qualità dei dati, riducendo il bias e selezionando un campione rappresentativo della popolazione.

Il campionamento casuale è un metodo di campionamento in cui ogni membro disponibile della popolazione ha la stessa probabilità di essere scelto per il campione. Il campionamento casuale può ridurre il bias e fornire una stima più accurata della popolazione.

## **Bias di selezione**

l bias di selezione si verifica quando i dati vengono selezionati in modo non rappresentativo della popolazione da cui provengono. Oppure si scelgano i dati che confermano la nostra ipotesi, tralasciando quelli che la contraddicono. Questo può portare a conclusioni fuorvianti o effimere.

Ecco alcuni esempi di bias di selezione:

- Campionamento non casuale: Se si sceglie un campione di dati in modo non casuale, è probabile che il campione non sia rappresentativo della popolazione.
- Cherry-picking: Se si scelgono solo i dati che supportano la propria ipotesi, è probabile che si arrivi a conclusioni errate.
- Selezione di intervalli di tempo: Se si esaminano solo i dati di un periodo di tempo specifico, è possibile che si trascurino fattori importanti che potrebbero influenzare i risultati.
- Interruzione di un esperimento: Se si interrompe un esperimento prima della fine, è possibile che si arrivi a conclusioni errate.

Per evitare il bias di selezione, è importante seguire le seguenti linee guida:

1. Utilizzare un campionamento casuale quando possibile.
2. Valutare tutti i dati disponibili, non solo quelli che supportano la propria ipotesi.
3. Considerare tutti i fattori che potrebbero influenzare i risultati.
4. Condurre l'esperimento fino alla fine.

## **Tipi di campionamento**

I possibili tipi di campionamento possono essere classificati in due categorie principali: campionamento probabilistico e campionamento non probabilistico.

### **Campionamento probabilistico**

Nel campionamento probabilistico, ogni elemento della popolazione ha la stessa probabilità di essere selezionato per il campione. I diversi metodi di campionamento probabilistico includono:

* **Campionamento casuale semplice:** Ogni elemento della popolazione ha la stessa probabilità di essere selezionato, indipendentemente da qualsiasi altra informazione.

* **Campionamento stratificato:** La popolazione viene suddivisa in strati, e poi un campione casuale viene estratto da ciascun strato. Questo metodo assicura che tutti gli strati della popolazione siano rappresentati nel campione.

* **Campionamento a grappoli:** La popolazione viene suddivisa in grappoli, e poi un campione di grappoli viene selezionato. I membri di ciascun grappolo vengono quindi inclusi nel campione. Questo metodo è efficiente quando è difficile o costoso accedere a tutti gli elementi della popolazione.

* **Campionamento sistematico:** Un elemento viene selezionato casualmente e poi ogni elemento successivo viene selezionato a intervalli regolari.




### **Campionamento non probabilistico**

Nel campionamento non probabilistico, non tutti gli elementi della popolazione hanno la stessa probabilità di essere selezionati per il campione. I diversi metodi di campionamento non probabilistico includono:

* **Campionamento di convenienza:** I dati vengono raccolti da elementi della popolazione che sono facilmente accessibili o disponibili.

* **Campionamento a caso:** Gli elementi della popolazione vengono selezionati casualmente, ma non in modo uniforme.

* **Campionamento per quote:** La popolazione viene suddivisa in categorie, e poi un numero di elementi viene selezionato da ciascuna categoria per garantire che tutte le categorie siano rappresentate nel campione.

* **Campionamento direzionale:** I dati vengono raccolti da elementi della popolazione che sono ritenuti rappresentativi della popolazione in generale.

# **La distribuzione campionaria di una statistica**

Il termine "distribuzione campionaria di una statistica" si riferisce alla distribuzione di una statistica campione su molti campioni estratti dalla stessa popolazione. Gran parte della statistica classica si occupa di trarre inferenze da (piccoli) campioni a (molto grandi) popolazioni.

Tipicamente, un campione viene estratto con l'obiettivo di misurare qualcosa (con una statistica campionaria) o modellare qualcosa (con un modello statistico o di apprendimento automatico).

Dato che la nostra stima o modello si basa su un campione, potrebbe essere in errore; potrebbe essere diverso se estraiamo un campione diverso. Siamo quindi interessati a quanto potrebbe essere diverso, una preoccupazione chiave è la variabilità campionaria.

Se avessimo molti dati, potremmo estrarre ulteriori campioni e osservare direttamente la distribuzione di una statistica campionaria. In genere, calcoleremo la nostra stima o modello utilizzando tutti i dati facilmente disponibili, quindi l'opzione di estrarre ulteriori campioni dalla popolazione non è prontamente disponibile.

La distribuzione di una statistica campione, come la media, è probabilmente più regolare e a campana rispetto alla distribuzione dei dati stessi. Più grande è il campione su cui si basa la statistica, più è vero. Inoltre, più grande è il campione, più ristretta è la distribuzione della statistica campione.

## **Teorema del limite centrale**

Il fenomeno che abbiamo appena descritto è definito teorema del limite centrale. Dice che le medie estratte da più campioni assomiglieranno alla familiare curva a campana, anche se la popolazione di origine non è distribuita normalmente, purché la dimensione del campione sia sufficientemente grande e la deviazione dei dati dalla normalità non sia troppo grande.

Immagina di avere una popolazione di dati, ad esempio la distribuzione delle altezze delle persone in un paese. Se estraiamo un campione da questa popolazione, la media del campione sarà una stima della media della popolazione. Tuttavia, la media del campione non sarà necessariamente uguale alla media della popolazione.

Il teorema del limite centrale afferma che, se la dimensione del campione è sufficientemente grande, la distribuzione delle medie campionarie si avvicina sempre più a una distribuzione normale, anche se la distribuzione della popolazione non è normale.

In altre parole, se estraiamo un numero sufficientemente grande di campioni dalla stessa popolazione, le medie dei campioni saranno distribuite in modo simile a una campana, con una media di μ e una deviazione standard di σ/√n.

Il teorema del limite centrale è un risultato importante in statistica perché ci consente di fare inferenze sulla popolazione sulla base di un campione. Ad esempio, possiamo utilizzare il teorema del limite centrale per calcolare intervalli di confidenza per la media della popolazione o per testare ipotesi sulla media della popolazione.

```python
# Definiamo la distribuzione della popolazione
mu = 100
sigma = 15

# Generiamo un campione di dimensioni 100
data = np.random.normal(mu, sigma, 100)

# Calcoliamo la media del campione
sample_mean = np.mean(data)

# Calcoliamo la deviazione standard del campione
sample_std = np.std(data)

# Visualizziamo la distribuzione della popolazione
plt.hist(data, bins=50)
plt.title("Distribuzione della popolazione")
plt.show()

# Visualizziamo la distribuzione delle medie campionarie
sample_means = []
for i in range(1000):
    sample = np.random.normal(mu, sigma, 100)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)

plt.hist(sample_means, bins=50)
plt.title("Distribuzione delle medie campionarie")
plt.show()
```

## **Standard error**


Lo standard error è una misura della variabilità delle stime della statistica di interesse. In altre parole, ci dice quanto possiamo aspettarci che l'estimate della statistica di interesse vari da campione a campione.

Per esempio, se estraiamo un campione di 100 persone dalla popolazione generale e calcoliamo la media del campione, la media del campione sarà una stima della media della popolazione. Tuttavia, è probabile che la media del campione sia diversa dalla media della popolazione. Lo standard error ci dice quanto possiamo aspettarci che la media del campione vari da campione a campione.

Lo standard error è calcolato come segue:

**SE = s / √n**

dove:

- SE è lo standard error
- s è la deviazione standard della popolazione
- n è la dimensione del campione

Come si può vedere dalla formula, lo standard error diminuisce all'aumentare della dimensione del campione. Questo perché, con un campione più grande, abbiamo più dati su cui basarci per stimare la statistica di interesse.


## **Bootstrap**

Il metodo Bootstrap è una tecnica statistica che viene utilizzata per stimare la distribuzione campionaria di una statistica. Il metodo funziona ri-campionando i dati del campione originale con reimmissione.

image.png

Per esempio, se estraiamo un campione di 100 persone dalla popolazione generale e calcoliamo la media del campione, la media del campione sarà una stima della media della popolazione. Il metodo Bootstrap ci consente di stimare la distribuzione campionaria della media del campione, che ci dice quanto possiamo aspettarci che la media del campione vari da campione a campione.

Il metodo Bootstrap funziona come segue:

1. Estraiamo un campione di dimensioni n dai dati del campione originale, con reimmissione.
2. Calcoliamo la statistica di interesse sul campione ri-campionato.
3. Ripetiamo i passaggi 1 e 2 un numero elevato di volte.
4. La distribuzione delle statistiche calcolate è la distribuzione campionaria della statistica di interesse.

In pratica, il metodo Bootstrap viene spesso utilizzato per stimare lo standard error di una statistica. In questo caso, la distribuzione campionaria della statistica di interesse viene utilizzata per calcolare la deviazione standard della distribuzione.

Il bootstrap serve a valutare l'errore standard perché fornisce una stima della variabilità della statistica di interesse da campione a campione.

In particolare, il bootstrap stima la distribuzione campionaria della statistica di interesse ri-campionando i dati del campione originale con reimmissione. La deviazione standard della distribuzione campionaria è quindi utilizzata come stima dell'errore standard.

**Il bootstrap non compensa un piccolo campione; non crea nuovi dati, né riempie le lacune in un set di dati esistente. Ci informa semplicemente su come si comporterebbero molti campioni aggiuntivi se tirati da una popolazione simile al nostro campione originale**

**esempio**: Questo codice genera un campione di 100 dati da una distribuzione normale con media 100 e deviazione standard 15. Quindi, esegue il metodo Bootstrap 1000 volte, ri-campionando i dati del campione originale con reimmissione prendendo solo 50 elementi. Infine, calcola lo standard error bootstrap come la deviazione standard della distribuzione delle medie campionarie bootstrap.

```python
# Definiamo i dati del campione
data = np.random.normal(100, 15, 100)

# Calcoliamo la media del campione
sample_mean = np.mean(data)

# Eseguiamo il metodo Bootstrap
bootstrap_samples = []
for i in range(1000):
    bootstrap_sample = np.random.choice(data, size=50, replace=True)
    bootstrap_samples.append(np.mean(bootstrap_sample))

# Calcoliamo lo standard error bootstrap
bootstrap_std = np.std(bootstrap_samples)

print("Media del campione:", sample_mean)
print("Standard error bootstrap:", bootstrap_std)
```

 il metodo bootstrap funziona ricampionando il campione originale un certo numero di volte, con reimmissione. Per ogni ricampionamento, si calcola la statistica di interesse. La distribuzione delle statistiche calcolate sui ricampionamenti approssima la distribuzione campionaria della statistica nella popolazione originale.

Quindi, il metodo bootstrap può essere utilizzato per stimare la media, la varianza, gli intervalli di confidenza e i p-value di una statistica, anche quando non si conosce la distribuzione della popolazione originale.

Ad esempio, se si ha un campione di dati sulla lunghezza di un certo tipo di albero, e si vuole stimare la media della popolazione di alberi, si può utilizzare il metodo bootstrap. Si ricampiona il campione un certo numero di volte, con reimmissione. Per ogni ricampionamento, si calcola la media del campione. La distribuzione delle medie calcolate sui ricampionamenti approssima la distribuzione campionaria della media nella popolazione di alberi.

## **Intervallo di confidenza**

un intervallo di confidenza è un intervallo di valori che è probabile contenga il vero valore di un parametro di una popolazione. Il parametro può essere la media, la proporzione o qualsiasi altra quantità che si desidera stimare.
Si tratta quindi di un metodo statistico per stimare il valore di una popolazione, basandosi sui dati di un campione. In particolare, l'IC fornisce un intervallo di valori all'interno del quale si trova, con una certa probabilità, il valore reale della popolazione.

Per calcolare un intervallo di confidenza, è necessario conoscere la stima del valore della popolazione, il valore critico della distribuzione corrispondente e l'errore standard.

La stima del valore della popolazione è il valore che si ottiene dai dati del campione. Ad esempio, se si vuole stimare la media della popolazione, la stima sarà la media del campione.

Il valore critico della distribuzione corrisponde al livello di confidenza desiderato. Ad esempio, se si vuole un intervallo di confidenza al 95%, il valore critico sarà zα/2, dove zα è il quantile z della distribuzione corrispondente al livello di confidenza (1 - α)/2.

L'errore standard è un'indicazione della precisione della stima. Più piccolo è l'errore standard, più precisa è la stima.

**IC = stima ± valore critico ∙ errore standard**

Ad esempio, supponiamo di voler stimare la media altezza degli italiani. Possiamo raccogliere un campione di italiani e calcolare la media altezza del campione. Tuttavia, questo valore non è necessariamente uguale alla media altezza della popolazione. Se raccogliessimo un altro campione, potremmo ottenere una media altezza diversa.

Un intervallo di confidenza ci dice quanto siamo sicuri che il vero valore del parametro si trovi all'interno dell'intervallo. Ad esempio, possiamo calcolare un intervallo di confidenza al 95% per la media altezza degli italiani. Questo significa che se ripetiamo il processo di campionamento 100 volte, ci aspettiamo che il vero valore della media altezza si trovi all'interno dell'intervallo calcolato 95 volte.

In pratica, il livello di confidenza si basa su un'assunzione sulla distribuzione del parametro della popolazione. Se si assume che la distribuzione del parametro della popolazione sia normale, si può utilizzare la distribuzione normale per calcolare gli intervalli di confidenza. Se la distribuzione del parametro della popolazione non è normale, è necessario utilizzare un'altra distribuzione, come la distribuzione t di Student.

*esempio popolazione normale*

```python
# Dati del campione
altezza = [1.70, 1.65, 1.80, 1.75, 1.68]

# Calcolo della media del campione
media = np.mean(altezza)

# Calcolo dell'intervallo di confidenza al 95%
intervallo_confidenza = np.percentile(altezza, [2.5, 97.5])

# Stampa dell'intervallo di confidenza
print(intervallo_confidenza)
```

*esempio di t di Student*


```python
# Dati del campione
altezza = [1.70, 1.65, 1.80, 1.75, 1.68]
# Calcolo della media del campione
media = np.mean(altezza)
# Calcolo dell'intervallo di confidenza al 95%
intervallo_confidenza = stats.t.interval(0.95, len(altezza) - 1, loc=media, scale=stats.sem(altezza))
# Stampa dell'intervallo di confidenza
print(intervallo_confidenza)
```

Utilizziamo la t di Student se una popolazione non è ritenuta normale perché è una distribuzione approssimativa della distribuzione normale quando la dimensione del campione è piccola.

Quando la dimensione del campione è grande, la distribuzione t di Student si avvicina alla distribuzione normale. Tuttavia, quando la dimensione del campione è piccola, la distribuzione t di Student è più piatta della distribuzione normale. Questo significa che la distribuzione t di Student è più conservatrice, ovvero fornisce intervalli di confidenza più larghi.

In generale, si consiglia di utilizzare la distribuzione t di Student quando la dimensione del campione è inferiore a 30. Se la dimensione del campione è superiore a 30, è possibile utilizzare la distribuzione normale, a meno che non ci siano prove che la distribuzione della popolazione non sia normale.
